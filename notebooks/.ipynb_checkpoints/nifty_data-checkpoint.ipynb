```python
from nsepy import get_history
from datetime import date
# sbin = get_history(symbol='SBIN',
#                    start=date(2015,1,1),
#                    end=date(2015,1,10))
```

```python
def get_date_from_str(str_date):
    '''
    str_date format: 'yyyy-mm-dd'
    '''
    return date(int(str_date.split('-')[0]), int(str_date.split('-')[1]), int(str_date.split('-')[2]))
```

```python
get_date_from_str(TRAIN_START_DATE)
```

```python
get_date_from_str(TRADE_2_END_DATE)
```

```python
get_history(symbol='NIFTY',
           start=get_date_from_str(TRAIN_START_DATE),
           end=get_date_from_str(TRADE_2_END_DATE),
           index=True)
```
